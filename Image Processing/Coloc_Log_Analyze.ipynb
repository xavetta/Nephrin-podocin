```python
import pandas as pd
import numpy as np
```

```python
def log_kiertekeles(logfile):
    headlines = [logfile.index(row) for row in logfile if 'Coloc_Job_Name' in row]
    headlines.sort()
    filenames = [' '.join(' '.join(' '.join(logfile[row].split(' ')[1:]).split('versus')[:1]).split('_')[4:-1]) for row in headlines]
    mutations = set([(name.split(' ')[1]).split('-')[0] for name in filenames])
    values = []
    for index in headlines:
        if headlines.index(index) < len(headlines)-1:
            index2 = headlines[headlines.index(index)+1]
        else:
            index2 = len(logfile)
        relevantrow = [row for row in logfile[index:index2] if "Pearson's R value (above threshold)" in row][0].split(' ')[-1].split('\n')[0]
        values.append(float(relevantrow))
    Masterlist = list(zip(filenames, values))
    analysis = pd.DataFrame(Masterlist, columns = ['Filename', 'P value'])
    analysis = analysis.sort_values(by=['Filename'])
    analysis.to_excel('Pearson_values.xlsx', index=False)
    return analysis, mutations
```

```python
with open ('Log.txt','r') as file:
    data = file.readlines()
table, mutations = log_kiertekeles(data)
display(table)
```

```python
for mut in mutations:
    values = [table['P value'][row] for row in range(len(table)) if mut in table['Filename'][row]]
    print(f'{mut}: Q1 = {np.round(np.percentile(values, 25),3)}, Median = {np.round(np.median(values),3)}, Q3 = {np.round(np.percentile(values, 75),3)},')
```
